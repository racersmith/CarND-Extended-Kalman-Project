# CarND Extended Kalman Filter
Generate visualization of postion and velocity estimates from the Kalman filter in comparison to the ground truth.

```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

## Define Data

```python
data_labels = ("est_px", "est_py", "est_vx", "est_vy", "meas_px", "meas_py", "gt_px", "gt_py", "gt_vx", "gt_vy")
data_files = ('out-1.txt', 'out-2.txt', 'out-3.txt')
```

## Generate Plots

```python
for file in data_files:
    print(file)
    data = np.genfromtxt(file, names = data_labels)
    plt.figure(figsize=(20, 8))
    plt.axis('equal')
    
    plt.subplot(1,2,1)
    plt.plot(data['gt_px'], data['gt_py'], 'g', alpha=0.7, linewidth=8, label='Actual Position')
    plt.plot(data['est_px'], data['est_py'], 'r.', label='Estimated Position')
    plt.xlim(min(data['gt_px']-2), max(data['gt_px'])+1)
    plt.legend(loc=2)
    plt.title("Position")

    plt.subplot(1,2,2)
    plt.plot(data['gt_vx'], data['gt_vy'], 'g', alpha=0.7, linewidth=1, label='Actual Velocity')
    plt.plot(data['est_vx'], data['est_vy'], 'r.', label='Estimated Velocity')
    plt.title("Velocity")
    plt.xlim(min(data['gt_vx']-1), max(data['gt_vx']+1))
    plt.legend(loc=2)
    plt.show()
```
